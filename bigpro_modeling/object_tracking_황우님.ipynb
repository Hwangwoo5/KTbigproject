```python
import cv2
from ultralytics import YOLO

# YOLOv8 모델 불러오기
model = YOLO('./runs/detect/train7/weights/best.pt') 
#새로 생성된 run/dectect 폴더 아래 중 가장 나중에 생성된 train뭐시기 폴더 있을거에요 그거 경로 설정해주세요.
#ex. ./runs/detect/train1/weights/best.pt로
# 동영상 파일 경로 설정
video_path = './blackbackbutterfly.mp4'
cap = cv2.VideoCapture(video_path)

# 윈도우 창의 이름 설정
cv2.namedWindow("YOLOv8 Inference", cv2.WINDOW_NORMAL)

# 창 크기 조절
cv2.resizeWindow("YOLOv8 Inference", 1000, 800)  # 너비 1000, 높이 800으로 조절

# 동영상 프레임을 반복 처리
while cap.isOpened():
    # 동영상에서 프레임 읽기
    success, frame = cap.read()

    if success:
        # 프레임에서 YOLOv8 추론 실행
        results = model(frame)

        # 결과를 프레임에 시각화
        annotated_frame = results[0].plot()

        # 주석이 달린 프레임을 디스플레이
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # 'q' 키를 누르면 루프를 종료
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # 동영상이 끝나면 루프를 종료
        break

# 비디오 캡처 객체 해제 및 디스플레이 창 닫기
cap.release()
cv2.destroyAllWindows()
```
